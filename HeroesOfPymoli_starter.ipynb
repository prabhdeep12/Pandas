### Heroes Of Pymoli Data Analysis
* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python

import pandas as pd
import numpy as np


file_to_load = "Resources/purchase_data.csv"


purchase_data = pd.read_csv(file_to_load)

df = purchase_data
df.head()

```

## Player Count

* Display the total number of players


```python
total_players = len(df["SN"].unique())
total_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
items = len(df["Item ID"].unique())
avg_price = df["Price"].mean()
purchase_count = df["Purchase ID"].count()
revenue = df["Price"].sum()

purchasing_analysis_df = pd.DataFrame({
   "Number of Unique Items": items,
   "Average Price": avg_price,
   "Number of Purchases": purchase_count,
   "Total Revenue": revenue
}, index=[0])

purchasing_analysis_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender = df["Gender"]
gender_count = df.groupby("Gender")["SN"].nunique()
gender_count

gender_percent = (gender_count/total_players)* 100
gender_percent

gender_demographics_df = pd.DataFrame({
   "Total Count": gender_count,
   "Percent of Players": gender_percent,
})

gender_demographics_df["Percent of Players"] = gender_demographics_df["Percent of Players"].map("{0:,.2f}%".format)

gender_demographics_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_count_g = df["Gender"].value_counts()
avg_purchase_price = df.groupby("Gender")["Price"].mean()
total_purchase_value = df.groupby("Gender")["Price"].sum()
avg_purchase_total_per_person = df.groupby("Gender")["Price"].mean()
purchase_count_unique = df.groupby("Gender")["SN"].nunique()

purchasing_analysis_df = pd.DataFrame({
   "Purchase Count": purchase_count_g,
   "Average Purcahse Price": avg_purchase_price,
   "Total Purchase Value": total_purchase_value,
   "Average Purchase Total per Person": total_purchase_value/purchase_count_unique
})

purchasing_analysis_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_bins_name = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", ">40"]
age_bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
df["Age Demographics"] = pd.cut(df["Age"], age_bins, labels=age_bins_name)

purchase_count_age = df.groupby("Age Demographics")["SN"].nunique()
avg_purchase_age = df.groupby("Age Demographics")["Price"].nunique().sum()
total_purchase_value_age = df.groupby("Age Demographics")["SN"].nunique().sum()
avg_per_person_age = df.groupby("Age Demographics")["Price"].mean()
age_count_g_unique = df.groupby("Age Demographics")["SN"].nunique()

age_demographics_df = pd.DataFrame({
   "Total Count": purchase_count_age,
   "Percentage of Players": purchase_count_age/total_purchase_value_age

})

age_demographics_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_count_a = df.groupby("Age Demographics")["Purchase ID"].count()
purchase_count_unique = df.groupby("Age Demographics")["SN"].nunique()
avg_purchase_price = df.groupby("Age Demographics")["Price"].mean()
total_purchase_value = df.groupby("Age Demographics")["Price"].sum()
avg_total_purchase_per_person = df.groupby("Age Demographics")["Price"].sum()/purchase_count_unique

purchasing_analysis_df = pd.DataFrame({
   "Purchase Count": purchase_count_a,
   "Average Purcahse Price": avg_purchase_price,
   "Total Purchase Value": total_purchase_value,
   "Avg Total Purchase per Person": total_purchase_value/purchase_count_unique
})

purchasing_analysis_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
user_total = purchase_data.groupby(["SN"]).sum()["Price"].rename("Total Purchase Value")
user_average = purchase_data.groupby(["SN"]).mean()["Price"].rename("Average Purchase Price")
user_count = purchase_data.groupby(["SN"]).count()["Price"].rename("Purchase Count")


user_data = pd.DataFrame({"Total Purchase Value": user_total, "Average Purchase Price": user_average, "Purchase Count": user_count})


user_data["Average Purchase Price"] = user_data["Average Purchase Price"].map("${:,.2f}".format)
user_data["Total Purchase Value"] = user_data["Total Purchase Value"].map("${:,.2f}".format)
user_data = user_data.loc[:,["Purchase Count", "Average Purchase Price", "Total Purchase Value"]]


user_data.sort_values("Total Purchase Value", ascending=False).head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python

item_data = purchase_data.loc[:,["Item ID", "Item Name", "Price"]]


total_item_purchase = item_data.groupby(["Item ID", "Item Name"]).sum()["Price"].rename("Total Purchase Value")
average_item_purchase = item_data.groupby(["Item ID", "Item Name"]).mean()["Price"]
item_count = item_data.groupby(["Item ID", "Item Name"]).count()["Price"].rename("Purchase Count")

item_data_pd = pd.DataFrame({"Total Purchase Value": total_item_purchase, "Item Price": average_item_purchase, "Purchase Count": item_count})
item_data_pd["Item Price"] = item_data_pd["Item Price"].map("${:,.2f}".format)
item_data_pd ["Purchase Count"] = item_data_pd["Purchase Count"].map("{:,}".format)
item_data_pd["Total Purchase Value"] = item_data_pd["Total Purchase Value"].map("${:,.2f}".format)
item_data_pd = item_data_pd.loc[:,["Purchase Count", "Item Price", "Total Purchase Value"]]


item_data_pd.sort_values("Purchase Count", ascending=False).head(5)
```
